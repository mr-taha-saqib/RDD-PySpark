```python
import pyspark
from pyspark import SparkContext, SparkConf
```

```python
conf = SparkConf().setAppName('appName').setMaster('local[*]')
sc=SparkContext(conf=conf)
```

```python
rdd1=sc.textFile("emp1.txt")
rdd2=sc.textFile("emp2.txt")
```

```python
rdd1.take(5)
```

```python
rdd2.take(5)
```

```python
rdd3=rdd1.join(rdd2)
```

```python
rdd3.collect()
```

```python
rdd3=rdd1.leftOuterJoin(rdd2)
```

```python
rdd3.collect()
```

```python
import csv
from io import StringIO

def loadRecords(line):

  input = StringIO.StringIO(line)
  reader = csv.DictReader(input,fieldnames=["emp_id","emp_salary","emp_experience"])
  return reader.next()
rdd11 = sc.textFile("emp1.txt").map(loadRecords)
rdd22 = sc.textFile("emp2.txt").map(loadRecords)
rdd33 = rdd11.join(rdd22)
rdd33.collect()
```
